```python
%matplotlib inline

from sklearn import datasets, cluster
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import scale
from sklearn.metrics import confusion_matrix,classification_report
```

# Convertir los campos a categoricos

```python
df_filtro_columnas_para_segmentacion["Prioridad"] = df_filtro_columnas_para_segmentacion["Prioridad"].astype('category')

df_filtro_columnas_para_segmentacion["Tipo Cotizacion a enviar"] = df_filtro_columnas_para_segmentacion["Tipo Cotizacion a enviar"].astype('category')

df_filtro_columnas_para_segmentacion["TipoUsuario"] = df_filtro_columnas_para_segmentacion["TipoUsuario"].astype('category')

df_filtro_columnas_para_segmentacion["SECTOR_FORMATEADO"] = df_filtro_columnas_para_segmentacion["SECTOR_FORMATEADO"].astype('category')
df_filtro_columnas_para_segmentacion.dtypes
```

```python
df_filtro_columnas_para_segmentacion["Prioridad_cat"] = df_filtro_columnas_para_segmentacion["Prioridad"].cat.codes
df_filtro_columnas_para_segmentacion["Tipo_Cotizacion_a_enviar_cat"] = df_filtro_columnas_para_segmentacion["Tipo Cotizacion a enviar"].cat.codes
df_filtro_columnas_para_segmentacion["TipoUsuario_cat"] = df_filtro_columnas_para_segmentacion["TipoUsuario"].cat.codes
df_filtro_columnas_para_segmentacion["SECTOR_FORMATEADO_cat"] = df_filtro_columnas_para_segmentacion["SECTOR_FORMATEADO"].cat.codes

df_filtro_columnas_para_segmentacion.head(2)
```

```python
#del df_filtro_columnas_para_segmentacion["Prioridad"]
#del df_filtro_columnas_para_segmentacion["Tipo Cotizacion a enviar"]
#del df_filtro_columnas_para_segmentacion["TipoUsuario"]
#del df_filtro_columnas_para_segmentacion["SECTOR_FORMATEADO"]
```

```python
df_filtro_columnas_para_segmentacion.head(2)
```

```python
df_tr =df_filtro_columnas_para_segmentacion

#Standardize
clmns = ['Prioridad_cat', 'Tipo_Cotizacion_a_enviar_cat','TipoUsuario_cat', 'SECTOR_FORMATEADO_cat']
df_tr_std = stats.zscore(df_tr[clmns])

#Cluster the data
kmeans = KMeans(n_clusters=4, random_state=0).fit(df_tr_std)
labels = kmeans.labels_

#Glue back to originaal data
df_tr['clusters'] = labels

#Add the column into our list
clmns.extend(['clusters'])

#Lets analyze the clusters
df_tr[clmns].groupby(['clusters']).mean()
```
