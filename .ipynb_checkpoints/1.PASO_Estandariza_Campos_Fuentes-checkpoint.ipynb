## Se estandariza la ciudad en el crm 

```python
crm_clientes['solo_ciudad'] =crm_clientes['ciudad'].str.split(' - ').str.get(0)
```

```python
del crm_clientes["ciudad"]
```

```python
crm_clientes.ciudad.unique()
```

```python
crm_clientes.rename(columns={'solo_ciudad': 'ciudad'}, inplace=True)
```

## Se estandariza la prioridad

```python
crm_clientes.Prioridad.unique()
```

```python
crm_clientes["Prioridad"]=crm_clientes.Prioridad.replace(['4 Baja'],['4_Baja']) 
```

## Se coloca el sector entregado por ITCloud

```python
crm_clientes_sector =  pd.merge(crm_clientes, sector_v1, how='left', left_on=['correoItcloud'],right_on=['correoItcloud'])
```

```python
crm_clientes_sector.rename(columns={'Sector_y': 'SECTOR_FORMATEADO'}, inplace=True)
```
